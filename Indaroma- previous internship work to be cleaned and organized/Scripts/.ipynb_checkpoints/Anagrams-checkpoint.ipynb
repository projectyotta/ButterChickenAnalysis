```python
# open the file containing the list of words 
# readme.md file says : The file "words" in this folder contains a list of words from Webster's Second International Dictionary (1934). The copyright has lapsed, and this file is included in Unix and OS X (and found at /usr/share/dict/words) as a reference wordlist for various uses. It is used in the Word Anagram Project (Chapter 3) in this course.
# open the file 
word = open('words','r')
word
```

```python
# read it line by line and see how the data looks like 
wordlist=word.readlines()
wordlist[:10]
```

```python
#get the total number of words in there 
len(wordlist)
```

```python
# data cleaning , two steps . Step 1 , remove the '/n' at the end of each line , and convert everything to lowercase 
print('Aaron\n'.strip())  # removes the \n 
print('Aaron'.lower()) # converts to lowercase 
```

```python
# do this for all the words 
wordclean = [word.strip().lower() for word in wordlist] # we are chaining the strip and lower functions 
wordclean[:10]
```

```python
# now , we notice that we have duplicates . One way to remove this is by using a set . 
wordunique=list(set(wordclean))
wordunique
# this seems to have worked , however we have lost the alphabetical order of the words . 
```

```python
# We need to sort it now . 
wordunique.sort()
wordunique
```

```python
# we can combine all of the parsing operations using comprehensions 
# let's look at this step by step 
'''     
first , open the file . open('words',r')
second , use strip and lower on it . word.strip().lower() for word in open('words',r')
third , put in in a set to remove duplicates set(word.strip().lower() for word in open('words',r'))
fourth , put that into a list . list(set(word.strip().lower() for word in open('words',r')))
fifth , sort that. sorted(list(set(word.strip().lower() for word in open('words',r'))))

'''

wordclean=sorted(list(set([word.strip().lower() for word in open('words','r')])))
wordclean
```

```python
# use sorted on a word to see how it works 
print(sorted('lives')) 
print(" ")
print(sorted('lives')==sorted('elvis')) # shoule be true since elvis and lives are anagrams 
print(" ")
print(sorted('dank')==sorted('memes')) # shoule be false since dank and memes are not anagrams 

```
